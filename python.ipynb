```python
#POO

#1 - Criar uma classe (um grupo de informações), sempre começa com a primeira letra MAIUSCULA
#2 - Criar um construtor, sempre comça com "def" e ele sempre terá no inicio o "__init__", depois passo as informações do construtor, a primeira é "self"
#3 - Criar os métodos
#4 - Criar as instancias da classe "casa1" e passa os atributos
5# - Executa os métodos "casa.mostrar()"
```

```python
class Casa:
    def __init__(self, cor, quartos, banheiros):
        self.cor = cor
        self.quartos = quartos
        self.banheiros = banheiros

    def mostrar_cor(self):
        print(f'A cor da casa é: {self.cor}')

    def mostrar_quartos(self):
        print(f'Esta casa tem: {self.quartos} quartos')

    def mostrar_banheiros(self):
        print(f'Esta casa tem: {self.banheiros} banheiros')

    def adicionar_quarto(self):
        self.quartos += 1
        print(f'Agora esta casa tem {self.quartos}')

    def pintar_casa(self, nova_cor):
        print(f'Pintando a casa de {self.cor} para {nova_cor}')

casa1 = Casa('Preta', 4, 2)
casa2 = Casa('Vermelha', 6, 3)

print('\nCasa 1: ')
casa1.mostrar_cor()
casa1.mostrar_quartos()
casa1.mostrar_banheiros()
casa1.adicionar_quarto()
casa1.pintar_casa('Ciano')

print('\nCasa 2: ')
casa2.mostrar_cor()
casa2.mostrar_quartos()
casa2.mostrar_banheiros()
casa2.adicionar_quarto()
```
